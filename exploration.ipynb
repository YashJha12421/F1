```python

results_orig = results.copy()
Path("debug_output").mkdir(exist_ok=True)
results_orig.to_csv("debug_output/results_backup.csv", index=False)

# helper: coerce position to numeric (non-numeric -> NaN)
results = results.copy()
results['position_num'] = pd.to_numeric(results['position'], errors='coerce')

# helper flags
results['has_valid_position'] = results['position_num'].notna()
# convert laps, points, positionOrder safely to numeric for sorting
results['laps_num'] = pd.to_numeric(results['laps'], errors='coerce').fillna(-1).astype(int)
results['points_num'] = pd.to_numeric(results['points'], errors='coerce').fillna(-999.0)
results['positionOrder_num'] = pd.to_numeric(results['positionOrder'], errors='coerce').fillna(9999).astype(int)

# preview problematic groups
dups = (results.groupby(['raceId','driverId']).size().reset_index(name='cnt').query('cnt>1'))
print("Number of duplicate (raceId,driverId) pairs:", dups.shape[0])
dups.head(10)

# define a "quality" sort order:
# 1) has_valid_position (True first)
# 2) laps_num (desc, more laps preferred)
# 3) points_num (desc)
# 4) positionOrder_num (asc; lower is better)
# 5) resultId (asc) as final deterministic tie-breaker
sort_cols = ['has_valid_position', 'laps_num', 'points_num', 'positionOrder_num', 'resultId']
results_sorted = results.sort_values(by=sort_cols, ascending=[False, False, False, True, True])

# pick best per group
results_best = results_sorted.drop_duplicates(subset=['raceId','driverId'], keep='first').reset_index(drop=True)

# rows dropped for inspection
dropped = results[~results.index.isin(results_best.index)].copy()
dropped.to_csv("debug_output/dropped_duplicate_rows.csv", index=False)
results_best.to_csv("debug_output/results_deduped.csv", index=False)

# Check uniqueness
print("duplicates after dedupe:", results_best.duplicated(subset=['raceId','driverId']).sum())
```

```python
def fn(a):
    if a>10:
        return 0
    if a==1:
        return 25
    if a ==2:
        return 18
    if a==3:
        return 15
    if a==4:
        return 12
    if a==5:
        return 10
    if a==6:
        return 8
    if a==7:
        return 6
    if a==8:
        return 4
    if a==9:
        return 2
    if a==10:
        return 1
    
    
```

```python
df = df.sort_values(['year', 'round'])
```

```python
team_race_points = (
    df
    .groupby(['year','round','constructorId'])
    .agg(
        team_points_race=('pointsAdjusted','sum'),
        n_drivers=('driverId','nunique')
    )
    .reset_index()
)

team_race_points['team_points_per_driver'] = (
    team_race_points['team_points_race'] / team_race_points['n_drivers']
)
```

```python
team_race_points = team_race_points.sort_values(['year','round'])

team_race_points['team_strength'] = (
    team_race_points
      .groupby(['year','constructorId'])['team_points_per_driver']
      .transform(lambda s: s.shift(1).expanding().mean())
)

team_race_points['team_strength'] = team_race_points['team_strength'].fillna(0)

```

```python
df = df.merge(
    team_race_points[['year', 'round', 'constructorId', 'team_strength']],
    on=['year', 'round', 'constructorId'],
    how='left'
)
df['team_strength'] = df['team_strength'].fillna(0)
```
